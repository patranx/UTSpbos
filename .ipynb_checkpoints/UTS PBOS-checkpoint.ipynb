```python
detik = int(input("Masukkan Jumlah Detik :"))

hari = detik/86400
temp1 = detik%86400
jam = temp1/3600
temp2 = temp1%3600
menit = temp2/60
detik = temp2%60

print ("Hari  : ",round(hari))
print ("Jam   : ",round(jam))
print ("Menit : ",round(menit))
print ("Detik : ",round(detik))

```
